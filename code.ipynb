```python
class HyperspectralCubeDataset(Dataset):
    def __init__(self, folder_path, target_size=(224, 32, 32)):
        self.paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.npy')]
        self.target_size = target_size

    def __len__(self):
        return len(self.paths)

    def center_crop(self, cube, target_shape):
        c, h, w = cube.shape
        tc, th, tw = target_shape

        assert c == tc, f"Expected {tc} channels, got {c}"

        # Crop center
        h_start = (h - th) // 2
        w_start = (w - tw) // 2
        return cube[:, h_start:h_start+th, w_start:w_start+tw]

    def __getitem__(self, idx):
        cube = np.load(self.paths[idx])  # shape: H x W x C
        cube = np.transpose(cube, (2, 0, 1)).astype(np.float32)  # to C x H x W

        if cube.shape[1] < self.target_size[1] or cube.shape[2] < self.target_size[2]:
            raise ValueError(f"ROI too small: got {cube.shape}, expected >= {self.target_size[1:]}")

        cube = self.center_crop(cube, self.target_size)
        cube = (cube - cube.mean()) / (cube.std() + 1e-5)

        # Dummy label for now
        label = 0.5
        return torch.tensor(cube), torch.tensor(label).float()
```

```python
class Hyperspectral3DCNN(nn.Module):
    def __init__(self, input_channels=224):
        super().__init__()
        self.conv_block = nn.Sequential(
            nn.Conv3d(1, 8, kernel_size=(7, 3, 3), stride=1, padding=1),
            nn.BatchNorm3d(8),
            nn.ReLU(),
            nn.MaxPool3d((2, 2, 2)),

            nn.Conv3d(8, 16, kernel_size=(5, 3, 3), padding=1),
            nn.BatchNorm3d(16),
            nn.ReLU(),
            nn.AdaptiveAvgPool3d(1)  # Output shape: [B, 16, 1, 1, 1]
        )

        self.regressor = nn.Sequential(
            nn.Flatten(),
            nn.Linear(16, 1)  # For regression task
        )

    def forward(self, x):
        # Input shape: [B, C, H, W] => reshape to [B, 1, C, H, W]
        x = x.unsqueeze(1)
        x = self.conv_block(x)
        return self.regressor(x).squeeze()
```

```python
# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load dataset
dataset = HyperspectralCubeDataset("kochia/kochia", target_size=(224, 32, 32))
loader = torch.utils.data.DataLoader(dataset, batch_size=4, shuffle=False)

# Initialize model
model = Hyperspectral3DCNN().to(device)
model.eval()

```
